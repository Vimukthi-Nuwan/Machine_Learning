# Logistic Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
df = pd.read_csv('Social_Network_Ads.csv')
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```
X_train
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
X_train
```

```
X_test
```

## Training the Logistic Regression model on the Training set

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting a new result

```
classifier.predict(sc.transform([[30, 87000]]))
```

## Predicting the Test set results

```
y_predict = classifier.predict(X_test)
result = np.concatenate((y_predict.reshape(len(y_predict), 1), y_test.reshape(len(y_test), 1)), 1)
print(f'{result}\n')
# Trying to find accuracy score manually (There is a method for find this. You can see it below)
true_count = 0
false_count = 0
for i in result:
  if i[0] == i[1]:
    true_count += 1
  else:
    false_count += 1

print(f"Count of correct prediction: {true_count}\nCount of wrong prediction: {false_count}\nPresentage of correct prediction (Accuracy score): {(true_count/(true_count + false_count))*100}")
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_predict)
accuracy_sc = accuracy_score(y_test, y_predict)
print(f'Confusion matrix:\n{cm}\n\nAccuracy score: {accuracy_sc}')
```

## Visualising the Training set results

```
from matplotlib.colors import ListedColormap

# Transform the training set back to original scale
X_set, y_set = sc.inverse_transform(X_train), y_train

# Create the grid for plotting
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=0.25),
    np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=0.25)
)

# Generate predictions for the grid and reshape for contour plotting
Z = classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape)

# Plot the decision boundary
plt.contourf(X1, X2, Z, alpha=0.75, cmap = ListedColormap(['orange', 'blue']))

# Set limits
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plot the training points
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(
        X_set[y_set == j, 0],  # Corrected variable name
        X_set[y_set == j, 1],  # Ensure this references the right data
        c=[ListedColormap(['orange', 'blue'])(i)],  # Fix the color assignment
        label=j
    )

# Add title and labels
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
